```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Amazon Sales data.csv")
```

```python
data.head()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Order Date']
```

```python
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month
```

```python
data['Year']
```

```python
sales_monthly = data.groupby(['Year', 'Month'])['Total Revenue'].sum()
sales_yearly = data.groupby(['Year'])['Total Revenue'].sum()
sales_yearly_monthly = data.groupby(['Year', 'Month'])['Total Revenue'].sum()
```

```python
plt.figure(figsize=(10, 5))
sales_monthly.plot(marker='o', color='b')
plt.title('Monthly Sales Trend')
plt.xlabel('Year-Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
